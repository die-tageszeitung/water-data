```python
import pandas as pd
import numpy as np
import sys
from datetime import datetime
```

```python
# import our own stuff
sys.path.append("../python/src")

from tools import read_water_data
```

```python
# set some options
#pd.set_option('display.max_rows', 600)
#pd.set_option('display.max_columns', 600)
#pd.set_option('display.width', None)
#pd.set_option('display.max_colwidth', None)

datadir = "../data/"
cachedir = "../data/cache" 
# the projectnumber '1' cannot be used as identification
# the analyse shows, that there are multible independent projects that use that number
ignoredprojectnumbers = ['n.a.','n.a','/','aggregate',np.nan,'1']
# topX of commitment,disbursment,grantequiv,count since we cannot order the different values
topX = 200 
# only projectes that have a commitment date after the 
activedate = datetime(year=2009,month=12,day=31)

# the relevant values are
relevantvalues=['USD_Commitment_Defl','USD_Disbursement_Defl','USD_GrantEquiv']
```

```python
%%time

# the data from oecd must be downloaded and saved as zips in the data-dir, check readme for links
oecddf = read_water_data(setname="fullset",datadir=datadir,cachedir=cachedir)

# tell me what it is
oecddf.describe()
```

```python
## show what amount of projects we ignored in general
tmp = oecddf[oecddf['ProjectNumber'].isin(ignoredprojectnumbers)]
display(tmp.describe())
```

```python
%%time
# related to oda grant, germany and waterprojects

# filter for flowcodes ODA Grants
#tmp = tmp[tmp['FlowName'].isin(['ODA Grants'])]
# filter for elements with values
tmp = tmp[tmp[relevantvalues].any(axis = 1)]
# filter for water sector
tmp = tmp[tmp['SectorCode']=='140']
# only projects related to germany as donor
tmp = tmp[tmp['DonorCode'] == "5"]
# only projects since 2010
tmp = tmp[tmp["CommitmentDate"] > datetime(year=2002,month=12,day=31)]

display(tmp.describe())
# what do we loss in total
display(pd.DataFrame(tmp[relevantvalues].sum()))
```

```python
# aggregate over german water projects and group
tmp = tmp.groupby(['ProjectTitle','DonorName','SectorName','PurposeName',
             'RecipientName',pd.Grouper(key="CommitmentDate",freq="y")
             
             
             ]).agg({
    "USD_Disbursement_Defl": np.sum,
    "USD_Commitment_Defl": np.sum,
    "USD_GrantEquiv": np.sum,
    "CommitmentDate": pd.Series.count,
    "AgencyName": lambda x: "; ".join(list(set(list(map(str,x))))), # concatinate involved agencys
    "ShortDescription": lambda x: "; ".join(list(set(list(map(str,x))))), # concat short descriptions
    "LongDescription": lambda x: "; ".join(list(set(list(map(str,x))))), # concatinate long descriptions
    'CrsID': lambda x: "; ".join(list(set(list(map(str,x))))) # concatinate involved CrsIDs
})
tmp.rename(columns={"CommitmentDate": "Elements"},inplace=True)
display(tmp)
```

```python
%%time
# filter for 'n.a.'-Projects and 'aggregate'
tmp = oecddf[~oecddf['ProjectNumber'].isin(ignoredprojectnumbers)]

# filter for flowcodes ODA Grants
#tmp = tmp[tmp['FlowName'].isin(['ODA Grants'])]

# filter for elements with values
tmp = tmp[tmp[relevantvalues].any(axis = 1)]

# create a list of projects
projectlist = list(np.unique(tmp.reset_index().set_index('ProjectNumber').index))
```

```python
%%time
# filter for projectnames
tmp = oecddf[oecddf['ProjectNumber'].isin(projectlist)]
#tmp = tmp[tmp['FlowName'].isin(['ODA Grants'])]
# only projects related to water
tmp = tmp[tmp['SectorCode']=='140']
# only projects related to germany as donor
tmp = tmp[tmp['DonorCode'] == "5"]
# only projects since 2010
tmp = tmp[tmp["CommitmentDate"] > activedate]
```

```python
%%time
# aggregate to get a list of biggest projects
# "big values" means high disbursement, commitment, 
# grantequiv or simply a high number of transactions

tmp = tmp.groupby(['ProjectNumber']).agg({"USD_Disbursement_Defl": np.sum,
                                          "USD_Commitment_Defl": np.sum,
                                          "USD_GrantEquiv": np.sum,
                                          "CommitmentDate": pd.Series.count})


a = list(tmp.reset_index().set_index('ProjectNumber').sort_values('USD_Disbursement_Defl',ascending=False).head(topX).index)
b = list(tmp.reset_index().set_index('ProjectNumber').sort_values('USD_Commitment_Defl',ascending=False).head(topX).index)
c = list(tmp.reset_index().set_index('ProjectNumber').sort_values('USD_GrantEquiv',ascending=False).head(topX).index)
d = list(tmp.reset_index().set_index('ProjectNumber').sort_values('CommitmentDate',ascending=False).head(topX).index)

topgermanwater = list(np.unique(a+b+c+d))
```

```python
# show what we loss  
#pd.set_option('display.max_rows', 600)
#pd.set_option('display.max_columns', 600)
tmp = oecddf[~oecddf['ProjectNumber'].isin(topgermanwater)].copy()
tmp = tmp[tmp['SectorCode']=='140']
# only projects related to germany as donor
tmp = tmp[tmp['DonorCode'] == "5"]
tmp = tmp[tmp["CommitmentDate"] > activedate]
display(tmp.describe(percentiles=[0.5,0.75,0.9,0.95,0.99]))

# aggregate over german water projects and group
tmp = tmp.groupby(["AgencyName",'FlowName',
                   #'RecipientName',
                   #pd.Grouper(key="CommitmentDate",freq="y"),
                   'PurposeName'
             ]).agg({
    "USD_Disbursement_Defl": np.sum,
    "USD_Commitment_Defl": np.sum,
    "USD_GrantEquiv": np.sum,
    "CommitmentDate": pd.Series.count,
    "ProjectNumber": lambda x: len(list(set(list(map(str,x))))), # count unique ProjectNumbers
    "ProjectTitle": lambda x: len(list(set(list(map(str,x))))) # count unique ProjectNumbers
})
tmp.rename(columns={
    "CommitmentDate": "Elements",
    "ProjectNumber": "Unique ProjectNumbers",
    "ProjectTitle": "Unique ProjectTitle"
},inplace=True)

display(tmp)
```
